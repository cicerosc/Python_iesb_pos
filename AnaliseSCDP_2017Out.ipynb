```python
# importar as bibliotecas necessárias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Manipulação dos dados

```python
# carregar os dados
df = pd.read_excel('../dados/201710_EmissaoPassagens_SCDP.xlsx',skiprows=3, skipfooter=3)
```

```python
# visualizar os dados iniciais
df.head()
```

```python
# visualizar os dados finais
df.tail()
```

```python
# visualizar a dimensão do conjunto de dados
df.shape
```

```python
# visualizar os metadados do arquivo importado
df.dtypes
```

```python
# corrigindo os nomes das colunas que apresentaram espaço
df.columns = df.columns.str.strip()
```

```python
# visualizar os metadados do arquivo imortado
df.info()
```

```python
# corrigindo as colunaas de data
## visualizar um resumo das datas
#df['Data Emissão Bilhete'].value_counts() #- não foi observado erro
#df['Data Embarque'].value_counts() #- não foi observado erro
## converter a data - não há data errada
df['Data Emissão Bilhete']= pd.to_datetime(df['Data Emissão Bilhete'])
df['Data Embarque']= pd.to_datetime(df['Data Embarque'])
# Visualizar as informações do data frame
df.info()
```

```python
# observar se as colunas com valores estão non formato correto
df.describe()
```

```python
# não foi observado nenhum erro quanto as colunas valor, contudo, as colunas de código precisa ser convertida para texto
```

```python
# corrigindo as colunas código para texto
df['Código do órgão superior'] = df['Código do órgão superior'].astype('str')
df['Código do órgão solicitante da viagem'] = df['Código do órgão solicitante da viagem'].astype('str')

df.info()
```

```python
# verificar quais anos foram emitidos os bilhetes
df['Data Emissão Bilhete'].dt.year.value_counts()
```

## Analise iniciais

```python
# observar algumas estatísticas
df.describe()
```

```python
# identificar qual foi o órgão que apresentou maior Valor Bilhete
df[df['Valor Bilhete']== df['Valor Bilhete'].max()].T
```

```python
# identificar quais forão os órgãos que apresentaram maior Valor Bilhete
df.nlargest(10,'Valor Bilhete')
```

```python
# identificar qual foi o órgão que apresentou menor Valor Bilhete
df[df['Valor Bilhete']== df['Valor Bilhete'].min()].T
```

```python
# Quantidade de bilhetes emitidos por órgão - lista com os 10 maiores
df.groupby('Nome do órgão superior')['Valor Bilhete'].count().nlargest(10)
```

```python
# Valor total dos bilhetes emitidos por órgão - lista com os 10 maiores
df.groupby('Nome do órgão superior')['Valor Bilhete'].sum().nlargest(10)
```

```python
# Quantidade de bilhetes emitidos por empresa aérea - lista com os 10 maiores
df.groupby('Companhia Aérea')['Valor Bilhete'].count().nlargest(10)
```

```python
# Valor total dos bilhetes emitidos por empresa aérea - lista com os 10 maiores
df.groupby('Companhia Aérea')['Valor Bilhete'].sum().nlargest(10)
```

```python
# Valor médio dos bilhetes emitidos por empresa aérea - lista com os 10 miores
df.groupby('Companhia Aérea')['Valor Bilhete'].mean().nlargest(10)
```

```python
# Percentual médio dos Desconto Aplicado por empresa aérea - lista com os 10 maiores
df.groupby('Companhia Aérea')['Percentual Desconto Aplicado'].mean().nlargest(10)
```

```python
# Percentual médio dos Desconto Aplicado por órgão - lista com os 10 maiores
df.groupby('Nome do órgão superior')['Percentual Desconto Aplicado'].mean().nlargest(10)
```

```python
# Adicionar a coluna tempestividade = diferença entre a data de embarque e de emissão
tempestividade = df['Data Embarque'] - df['Data Emissão Bilhete']
df['tempestividade'] = tempestividade.dt.days
df.info()
```

```python
# Verificar a media de tempo entre emissão do bilhete e a viagem 
#df['tempestividade'].mean()
df.groupby('Nome do órgão superior')['tempestividade'].mean().nlargest(10)
```

```python
# Quantidade de empresas utilizada por órgão - as 10 menores
df.groupby('Nome do órgão superior')['Companhia Aérea'].nunique().nsmallest(10)
```

### Gráficos

```python
# Distribuição dos valores dos bilhetes
plt.title("Valores dos Bilhetes das passagens governamenais x Valor Tarifa Comercial")
sns.distplot(df['Valor Tarifa Comercial'], label = "Valores das tarifas comerciais")
sns.distplot(df['Valor Bilhete'], label = "Valores dos Bilhetes")
```

```python
# Distribuição dos valores dos bilhetes
sns.set(rc={'figure.figsize':(15, 10)})
plt.title("Valores dos Bilhetes das passagens governamenais por companhia aérea")
sns.distplot(df[df['Companhia Aérea'].str.strip()=='MAP TRANSPORTES AEREOS LTDA - MAP']['Valor Bilhete'], label = "MAP TRANSPORTES AEREOS LTDA - MAP")
sns.distplot(df[df['Companhia Aérea'].str.strip()=='GOL Linhas Aéreas Inteligentes']['Valor Bilhete'], label = "GOL Linhas Aéreas Inteligentes")
sns.distplot(df[df['Companhia Aérea'].str.strip()=='AZUL LINHAS AÉREAS BRASILEIRA']['Valor Bilhete'], label = "AZUL LINHAS AÉREAS BRASILEIRA")
sns.distplot(df[df['Companhia Aérea'].str.strip()=='LATAM AIRLINES BRASIL']['Valor Bilhete'], label = "LATAM AIRLINES BRASIL")
sns.distplot(df[df['Companhia Aérea'].str.strip()=='Gol OU Latam - Parceria com Terceiro']['Valor Bilhete'], label = "Gol OU Latam - Parceria com Terceiro")
sns.distplot(df[df['Companhia Aérea'].str.strip()=='AVIANCA S.A.']['Valor Bilhete'], label = "AVIANCA S.A.")
plt.legend()
```

```python
# Gráfico da Quantidade de bilhetes emitidos por órgão - lista com os 10 maiores
df.groupby('Nome do órgão superior')['Valor Bilhete'].count().nlargest(10).plot.barh()
```

```python
# Gráfico Valor total dos bilhetes emitidos por órgão - lista com os 10 maiores
df.groupby('Nome do órgão superior')['Valor Bilhete'].sum().nlargest(10).plot.barh()
```

```python
# Gráfico da Quantidade de bilhetes emitidos por empresa aérea 
df.groupby('Companhia Aérea')['Valor Bilhete'].count().nlargest(10).plot.pie()
```

```python
# Gráfico do Valor total dos bilhetes emitidos por empresa aérea 
df.groupby('Companhia Aérea')['Valor Bilhete'].sum().plot.pie()
```

```python
# Gráfico box plot do Valor total dos bilhetes emitidos por empresa aérea 
sns.boxplot(x='Companhia Aérea', y='Valor Bilhete',data=df)
```

```python
# Verificando a relação entre a tempestividade e o valor do bilhete por companhia aérea
sns.pairplot(x_vars=['tempestividade'],y_vars=['Valor Bilhete'], data=df, size=12,hue='Companhia Aérea', kind='reg')
```

```python
# Verificando as correlações
corr = df[['Valor Bilhete','Valor Tarifa Comercial','Valor Tarifa Governo','tempestividade']].corr()
sns.heatmap(corr)
```
